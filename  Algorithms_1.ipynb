```python
#Initialization of libraries
import numpy as np
import timeit
import matplotlib.pyplot as plt
import random
```

## I. Exercises

```python
#Time spent on each N info gathering
def timespent(function, k, n):
    timeseries = []
    for y in range(n):
        t = 0
        for x in range(k):
            number = list(np.random.random(y))
            start = timeit.default_timer()
            function(number)
            end = timeit.default_timer()
            t += (end - start)
        timeseries.append(t / k)
    return timeseries
```

```python
#1. Constant function
def constant(input):
    return 0
```

```python
#2. The sum of elements
def summ_of_elements(input):
    sum = 0
    for i in input:
        sum += i
    return sum
```

```python
#3. The product of elements
def product(input):
    prod = 1
    for i in input:
        prod *= i
    return prod
```

```python
#4.1 Polynomial normal method
def polynominal_normal(input):
    x = 1.5
    result = 0
    for i in range(len(input)):
        result += input[i] * x ** i
    return result
```

```python
#4.2 Polynomial Horner method
def polynominal_horner(input): 
    x = 1.5
    result = 0
    for i in range(len(input), 0, -1):
        result = input[i - 1] + x * result
    return result
```

```python
#5. Bubble sort
def bubblesort(input):
    for z in range(len(input)):
        for x in range(len(input)-1):
            if input[x] > input[x + 1]:
                input[x], input[x + 1] = input[x + 1], input[x]
    return input
```

```python
#6. Quick sort
def QuickSort(A):
    if len(A) <= 1:   
        return A

    pivot = A[random.randrange(len(A))]  
    left = list(filter(lambda x: x < pivot, A))  
    center = [i for i in A if i == pivot]  
    right = list(filter(lambda x: x > pivot, A))  

    return QuickSort(left) + center + QuickSort(right) 
```

```python
#7. Timsort
def timsort(input):
    return np.sort(input, kind='stable')
```

## II. Generate random matrices 𝐴 and 𝐵 of size 𝑛 × 𝑛 with non-negative elements. Find the usual matrix product for 𝐴 and 𝐵.

```python
#Time spent on each N info gathering for matrices
def timespent_matrix(k, n):
    timeseries = []
    for y in range(1, n + 1):
        t = 0
        for x in range(k):
            matrix1 = np.random.random((y, y))
            matrix2 = np.random.random((y, y))
            start = timeit.default_timer()
            matrix1 @ matrix2
            end = timeit.default_timer()
            t += (end - start)
        timeseries.append(t / k)
    return timeseries
```

### Collecting time spent data

```python
#8. Product of matrices
ex_8 = timespent_matrix(5, 500)
```

### Matrices product visualization

```python
plt.figure(figsize=(10,5))
plt.plot(ex_8)
plt.ylabel('Time (seconds)')
plt.xlabel('N')
plt.legend(['Product of matrices'])
plt.show()
```
